```python
import cv2
import glob
import os
from google.colab.patches import cv2_imshow
```

```python
!unzip "/content/Bulai_1000.zip" 
```

```python
os.mkdir('/content/Bulai_gray_1000')
images_path = glob.glob('/content/Bulai_1000/*.jpg')
```

```python
i = 0
for image in images_path:
  img = cv2.imread(image)
  gray_images = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  # cv2_imshow( gray_images)
  cv2.imwrite('/content/Bulai_gray_1000/image%02i.jpg' %i, gray_images)
  i += 1
```

```python
!zip -r '/content/Bulai_gray_1000.zip' '/content/Bulai_gray_1000'
```

```python
os.mkdir('/content/Bulai_biner_1000')
```

```python
images_path = glob.glob('/content/Bulai_1000/*.jpg')
i = 0
for image in images_path:
  # im_gray = cv2.imread('grayscale_image.png', cv2.IMREAD_GRAYSCALE)
  img_gray = cv2.imread(image, cv2.IMREAD_GRAYSCALE)
  (thresh, img_bw) = cv2.threshold(img_gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
  # gray_images = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  # cv2_imshow( gray_images)
  cv2.imwrite('/content/Bulai_biner_1000/image%02i.jpg' %i, img_bw)
  i += 1
```
