```python
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
```

# **Membuat Histogram**

```python
img = cv2.imread('bulai-24.jpg',0)
img = cv2.resize(img, (200,400))
cv2_imshow(img)
# plt.imshow(img)
```

```python
histr = cv2.calcHist([img],[0],None,[256],[0,256])
```

```python
plt.plot(histr)
plt.show()
```

```python
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

Peningkatan Kecerahan

```python
bright_img = img + 100
cv2_imshow(bright_img)
```

```python
def compare_image (image1, image2, title):
  fig = plt.figure(title)
  plt.suptitle(title)
  # show first image
  ax = fig.add_subplot(1,2,1)
  plt.imshow(image1, cmap = plt.cm.gray)
  plt.axis("off")
	# show the second image
  ax = fig.add_subplot(1,2,2)
  plt.imshow(image2, cmap = plt.cm.gray)
  plt.axis("off")
  # show the images
  plt.show()
```

```python
compare_image(img, bright_img, "Peningkatan Kecerahan")
```

```python
plt.hist(bright_img.ravel(),256,[0,256])
plt.show()
```

Meregangkan Kontras

```python
contr_img = 2 * img
compare_image(img, contr_img, "Meregangkan Kontras")
```

```python
plt.hist(contr_img.ravel(),256,[0,256])
plt.show()
```

Kombinasi Kontras dan Kecerahan

```python
bright = img + 10
combine_img = bright * 2
compare_image(img, combine_img, "Kombinasi Kontras Kecerahan")
```

```python
plt.hist(combine_img.ravel(),256,[0,256])
plt.show()
```

Membalik Citra

```python
translated_img = 255 - img
compare_image(img, translated_img, "Membalik Citra")
```

```python
plt.hist(translated_img.ravel(),256,[0,256])
plt.show()
```
