```python
import cv2
from matplotlib import pyplot as plt
from google.colab.patches import cv2_imshow
import pandas as pd
import numpy as np
```

```python
#path untuk sample image
path = ('drive/MyDrive/Jagung All/Bulai_1000/')
```

```python
#fungsi untuk compare image
def compare_image (image1, image2, title):
  fig = plt.figure(title)
  plt.suptitle(title)
  # show first image
  ax = fig.add_subplot(1,2,1)
  plt.imshow(image1, cmap = plt.cm.gray)
  plt.axis("off")
	# show the second image
  ax = fig.add_subplot(1,2,2)
  plt.imshow(image2, cmap = plt.cm.gray)
  plt.axis("off")
  # show the images
  plt.show()
```

```python
#image asli
img_real = cv2.imread(path +'Bulai_1 (10).png',0)
```

## Erosi dan Dilasi

```python
#matriks 5x5 sebagai kernel
kernel = np.ones((5,5), np.uint8)
```

```python
img_erosion = cv2.erode(img_real, kernel, iterations=1)
img_dilation = cv2.dilate(img_real, kernel, iterations=1)
```

```python
compare_image(img_real, img_erosion, "Real dan Erosi")
```

```python
compare_image(img_real, img_dilation, "Real dan Dilasi")
```

## Opening dan Closing

```python
img_opening = cv2.morphologyEx(img_real, cv2.MORPH_OPEN, kernel)
img_closing = cv2.morphologyEx(img_real, cv2.MORPH_CLOSE, kernel)
```

```python
compare_image(img_real, img_opening, "Real dan Opening")
```

```python
compare_image(img_real, img_closing, "Real dan Closing")
```
